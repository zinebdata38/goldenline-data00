```python
import pandas as pd
import matplotlib.pyplot as plt
from faker import Faker

# Initialiser Faker
fake = Faker()

# Générer des données aléatoires
def generate_data(num_entries):
    data = {
        'Nom': [fake.last_name() for _ in range(num_entries)],
        'Prénom': [fake.first_name() for _ in range(num_entries)],
        'Âge': [fake.random_int(min=18, max=80) for _ in range(num_entries)],
        'Ville': [fake.city() for _ in range(num_entries)]
    }
    return pd.DataFrame(data)

# Générer les données
num_entries = 10  # Par exemple, génère 10 entrées
df = generate_data(num_entries)

# Afficher les données
print(df)

# Créer un graphique
plt.figure(figsize=(10, 6))
plt.hist(df['Âge'], bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribution des âges')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Générateur de Noms et Prénoms

Cette application génère un tableau avec des noms, prénoms, âges et villes aléatoires. 
Elle affiche également un histogramme de la distribution des âges.
